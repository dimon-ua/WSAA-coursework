```python
import csv
```

```python
FILENAME= "data.csv"
DATADIR = ""

with open (DATADIR + FILENAME, "rt") as fp:
    reader = csv.reader(fp, delimiter=",")
    for line in reader:
        print (line) # list type
```

```python
with open (DATADIR + FILENAME, "rt") as fp:
    reader = csv.reader(fp, delimiter=",")
    linecount = 0
    for line in reader:
        if not linecount: # first row ie header row
            print (f"{line}\n-------------------")
        else: # all subsequent rows
            print (line)
        linecount += 1
```

```python
#a. Convert the string that is read into an integer

with open (DATADIR + FILENAME, "rt") as fp:
    reader = csv.reader(fp, delimiter=",")
    linecount = 0
    total = 0
    for line in reader:
        if not linecount: # first row ie header row
            pass
        else: # all subsequent rows
            total += int(line[1]) # why 1
        linecount += 1
    print (f"average is {total/(linecount-1)}") # why -1 ?
    # calc without header that;s why -1

```

```python
# b. Use the quote parameter to read in the numbers as floats

with open (DATADIR + FILENAME, "rt") as fp:
    reader = csv.reader(fp, delimiter=",", quoting=csv.QUOTE_NONNUMERIC)
    linecount = 0
    total = 0
    for line in reader:
        if not linecount: # first row ie header row
            pass
        else: # all subsequent rows
            total += line[1] # why 1
            linecount += 1

print (f"average is {total/(linecount-1)}") # why -1 ? to avoid error - ZeroDivisionError: division by zero
```

```python
# 5. The CVS file could of course have been read in as a Dictionary object Using DictReader()

with open (DATADIR + FILENAME, "rt") as fp:
    # DictReader сам понимает, что первая строка — это заголовки
    reader = csv.DictReader(fp, delimiter=",", quoting=csv.QUOTE_NONNUMERIC)
    total = 0
    count = 0
    for line in reader:
        # Теперь мы обращаемся к данным по имени колонки 'age'
        total += line['age']
        # print (line)
        count += 1

print (f"average is {total/(count)}") # why is there no -1 this time?
# Beacuse DictReader starts count from second row and passes 1st row with id
```
