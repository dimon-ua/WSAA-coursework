```python
import requests
import urllib.parse
from config import apikeys as cfg
import json
```

```python
# https://html2pdf.app/documentation/#Python

url = "https://andrewbeatty1.pythonanywhere.com/bookviewer.html"
apiKey = cfg["htmltopdfkey"]
linkRequests = "https://api.html2pdf.app/v1/generate?html={0}&apiKey={1}".format(url, apiKey)

result = requests.get(linkRequests).content

with open("document.pdf", "wb") as handler:
    handler.write(result)
```

```python
filename = "repo_info.json"

url = 'https://api.github.com/repos/dimon-ua/aprivateone'

response = requests.get(url, auth=('token',cfg["githubkey"]))
repoJSON = response.json()

#print (response.json())
with open(filename, 'w') as fp:
    json.dump(repoJSON, fp, indent=4)
```
